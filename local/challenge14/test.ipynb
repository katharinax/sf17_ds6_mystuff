```python
cursor = hmm.aggregate([
    {"$group": {
        "_id": "$year",
        "year_count": {"$sum": 1} 
    }}
]) #do we need to know this?!
```

```python
years = []
counts = []
for doc in cursor:
    years.append(doc["_id"])
    counts.append(doc["year_count"])
```

```python
for i in range(len(years)):
    if years[i] == 1980:
        print(years[i], counts[i])
```
