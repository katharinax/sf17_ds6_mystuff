```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy
%matplotlib inline
```

```python
df = pd.read_csv("Practice_data.csv")
```

```python
df.tail()
```

```python
f, ax = plt.subplots(3, 2, figsize=(30, 10))
plt.tight_layout(pad = 5)
ax[0, 0].plot(df['Ind_Variable_1'], df['Dep_Variable'], 'co', markersize = 2)
ax[0, 0].set_title("Ind_Variable_1")

ax[0, 1].plot(df['Ind_Variable_2'], df['Dep_Variable'], 'co', markersize = 2)
ax[0, 1].set_title("Ind_Variable_2")

ax[1, 0].plot(df['Ind_Variable_3'], df['Dep_Variable'], 'co', markersize = 2)
ax[1, 0].set_title("Ind_Variable_3")

ax[1, 1].plot(df['Ind_Variable_4'], df['Dep_Variable'], 'co', markersize = 2)
ax[1, 1].set_title("Ind_Variable_4")

ax[2, 0].plot(df['Ind_Variable_5'], df['Dep_Variable'], 'co', markersize = 2)
ax[2, 0].set_title("Ind_Variable_5")

ax[2, 1].plot(df['Ind_Variable_6'], df['Dep_Variable'], 'co', markersize = 2)
ax[2, 1].set_title("Ind_Variable_6")
```

We think Ind_Variable_4 is positively correlated with our outcome.

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1])
#train, test = train_test_split(df, train_size = 0.75, random_state = 1)
```

### model 1

```python
results3.summary()
```

We think there was a clear pattern from the exploratory plots, but then our models are poorly fitted (R-squared ~ 0.59). The positive sign of the Ind_Variable_4 coefficient indicates positive correlation with the outcome. The p-values of Ind_Variable_1-4 are close to 0, which means high certainty that they are related to the outcome. The high p-values of Ind_Variable_5-6 indicate that we are not sure if they are associated with the outcome.
