### compare: rating stars vs. latent grouping vs sentiment compound

```python
comp_df = book_review[['review_rating', 'review_compound']].merge( \
          review_latent[['PCA_grouping']], left_index = True, right_index = True)
```

```python
comp_df.tail(1)
```

```python
def re_w_na(x):
    try:
        num = int(re.sub(r"[^0-9]*", r"", x))
        return num
    except:
        return -1
```

```python
compare_mx = pd.crosstab(comp_df['review_rating'], comp_df['PCA_grouping'])
compare_mx
```

```python
comp_df['PCA_grouping_num'] = comp_df['PCA_grouping'].map(lambda x: re_w_na(x))
comp_df.tail(1)
```

```python
sns.heatmap(compare_mx)
```
