```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.utils.extmath import randomized_svd
from sklearn.decomposition import TruncatedSVD
from nltk.tokenize import word_tokenize, wordpunct_tokenize, WhitespaceTokenizer, \
                          RegexpTokenizer, sent_tokenize, MWETokenizer
from nltk.util import ngrams
from nltk.chunk import ne_chunk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import normalize, MinMaxScaler
from sklearn.cluster import KMeans
import gensim
import warnings
import logging
import pickle

%matplotlib inline
pd.options.display.max_columns = 200
```

```python
book_review = pd.read_pickle("cleaned_w_bkrvwnlp.pkl")
book_review.tail(1)
```

```python
book_df = book_review.drop(['userid', 'review_rating', 'review_date', 'review_desc', 'bk_desc', \
                            'bk_author', 'bk_title'], axis = 1)
book_df = book_df.groupby(['bookid']).agg(np.mean)
```

```python
cols = book_df.columns
rows = book_df.index
book_df = book_df.fillna(book_df.mean())
book_df = pd.DataFrame(MinMaxScaler().fit_transform(book_df), columns = cols, index = rows)
        # note: not sure why but normalize() doesn't perform well
print(book_df.shape)
book_df.tail(1)
```

```python
# book_df.mean()
```

### latent features

```python
n_components = 30
```

```python
tSVD_model = TruncatedSVD(n_components = n_components, random_state = 1, n_iter = 10)
tSVD_model.fit(book_df)
```

```python
var_ratio_df = pd.DataFrame(tSVD_model.explained_variance_ratio_)
var_ratio_df.columns = ['var_ratio']
var_ratio_df['cum_var_ratio'] = var_ratio_df['var_ratio'].cumsum()
var_ratio_df
```

```python
plt.figure(figsize = (16, 4))
plt.plot(var_ratio_df['var_ratio'], 'b-o')
plt.show()
```

```python
book_latent = pd.DataFrame(tSVD_model.transform(book_df))
book_latent.set_index(rows, inplace = True)
book_latent.tail(1)
```

```python
book_latent['PCA_grouping'] = book_latent.idxmax(axis = 1)
book_latent.tail(10)
```

```python
book_latent['PCA_grouping'].value_counts()
```

### K-means

```python
n_components = 4
```

```python
kclust = KMeans(n_clusters = n_components)
```

```python
# book_df.tail(1)
```

```python
book_kclust = pd.DataFrame(kclust.fit_predict(book_df))
book_kclust.columns = ['kmeans_grouping']
book_kclust.set_index(rows, inplace = True)
book_kclust.tail(1)
```

### compare groupings

```python
book_grps = book_latent.merge(book_kclust, left_index = True, right_index = True)
book_grps.tail(1)
```

```python
pd.crosstab(book_grps['PCA_grouping'], book_grps['kmeans_grouping'])
```
