### #4. Make one figure with (N=the number of MPAA ratings there are) subplots, and in each plot the release date vs the domestic total gross.

```python
gp = movie.groupby('mpaa')
f, axs = plt.subplots(2, 2, figsize = (12, 5), sharex = 'all')
plt.tight_layout(h_pad = 2)
for ax, key in zip(axs.flatten(), gp.groups.keys()):
    subset = movie.iloc[gp.groups[key]]
    ax.plot(subset['release_date'], subset['life_gross'], 'co', markersize = 4)
    ax.set_title(key)
plt.show()
```

### #5. What director in your dataset has the highest gross per movie?

```python
movie_director = movie[movie['role'] == 'Director']
dir_gross = pd.DataFrame(movie_director.groupby('playerid').mean()['life_gross'])
max_name = dir_gross.sort_values(by = 'life_gross', ascending = False).iloc[0].name
```

### #6. Bin your dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean.

```python
movie['yr'] = movie['release_date'].dt.year
movie['month'] = movie['release_date'].dt.month
movie13 = movie[(movie['yr'] == 2013) & ~(np.isnan(movie['life_gross']))]
movie13 = movie13.reset_index(drop = True)
```

```python
gross_by_mon2013 = pd.DataFrame(movie13.loc[:, ['month', 'life_gross']].groupby('month').agg(['count', 'mean', 'std']))
gross_by_mon2013.columns = gross_by_mon2013.columns.droplevel()
gross_by_mon2013 = gross_by_mon2013.reset_index()
gross_by_mon2013['se'] = gross_by_mon2013['std'] / gross_by_mon2013['count'] ** 0.5
```

```python
#gross_by_mon2013
```

```python
x = gross_by_mon2013['month']
y = gross_by_mon2013['mean']
yerr = gross_by_mon2013['se']
plt.bar(x, y, yerr = yerr)
# plt.errorbar(x, y, yerr=yerr)
plt.title("Mean Domestic Total Gross by Month in 2013\n", fontweight = 'bold')
```
